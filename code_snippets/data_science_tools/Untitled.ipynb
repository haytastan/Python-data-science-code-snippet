```python
from distfit import distfit
import numpy as np

X = np.random.normal(0, 2, 1000)
y = [-8,-6,0,1,2,3,4,5,6]

# Initialize model
dist = distfit()

# Find best theoretical distribution for empirical data X
dist.fit_transform(X)
dist.plot()
```

```python
# Make prediction
dist.predict(y)
dist.plot()
```
